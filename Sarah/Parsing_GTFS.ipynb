```python
##create a list of the timestamps and corresponding URLs

from datetime import date, datetime, timedelta

def datetime_range(start, end, delta):
    current = start
    if not isinstance(delta, timedelta):
        delta = timedelta(**delta)
    while current < end:
        yield current
        current += delta
        
start = datetime(2017,6,1,0,1)
end = datetime(2017,7,1)

# files are created in five minute increments: 01, 06, 11, 16, 21, 26, 31, 36, 41, 46, 51, and 56 
for dt in datetime_range(start, end, {'days': 0, 'hours':0, 'minutes':5}):
    with open("June_dates.csv", "a") as output:
        writer = csv.writer(output, delimiter = " ")
        date = str(dt.strftime('%Y-%m-%d-%H-%M'))
        file_name = 'https://datamine-history.s3.amazonaws.com/gtfs-'+date
        writer.writerow([file_name])
output.close()
            
    #print (dt.strftime('%Y-%m-%d-%H-%M'))
    #print dt
    
```
